```python
#!pip install delayed

import pandas as pd
input_file = open("indian_diabetes.csv", encoding="ISO-8859-1")
df = pd.read_csv(input_file, header = 0, delimiter=",")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.sample(frac=1)
```

```python
len(df)
```

```python
train = df.iloc[:int(len(df)*0.7),:]
test = df.iloc[int(len(df)*0.7):int(len(df)*0.9),:]
dev = df.iloc[int(len(df)*0.9):,:]

print(len(train))
print(len(test))
print(len(dev))


print(df.columns)
```

```python
from sklearn.linear_model import Lasso 
max_r2=0
alpha_with_max_R2 = 0
import numpy as np
from sklearn.metrics import r2_score


vars = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']



alphas = np.logspace(-10, 1, 20)
#alphas = [0.003359818286283788]
index_of_max_alpha = 0
for a in alphas:
    lassoreg = Lasso(alpha=a,normalize=True, max_iter=1e5)
    lassoreg.fit(train[vars],train["Outcome"])

    y_pred = lassoreg.predict(test[vars])
    r2 = r2_score(test["Outcome"], y_pred)
    #adjusted_r2 = 1 - (((1 - r2) * (len(df) - 1)) / (len(df) - len(union) - 1))

    if max_r2 <= r2:
      max_r2 = r2
      alpha_with_max_R2 = a
      df_coeff = pd.DataFrame(vars,columns=['Variables'])

      coeffs= pd.DataFrame(lassoreg.coef_,columns=["Coefficients"])
      df_coeff = df_coeff.merge(coeffs,how='inner',left_index=True,right_index=True)

print("The highest r2 sccore was: "+str(max_r2)+" with alpha "+str(alpha_with_max_R2))
df_coeff
```
